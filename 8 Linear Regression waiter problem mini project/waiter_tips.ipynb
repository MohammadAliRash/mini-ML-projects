```python
from sklearn.preprocessing import LabelEncoder

encoded_sex = LabelEncoder().fit(['Male', 'Female']).transform(data.sex)
encoded_smoker = LabelEncoder().fit(['Yes', 'No']).transform(data.smoker)
encoded_day = LabelEncoder().fit(["Thur", "Fri", "Sat", "Sun"]).transform(data.day)
encoded_time = LabelEncoder().fit(['Lunch', 'Dinner']).transform(data.time)
```

```python
data2 = pd.read_csv('tips.csv')
data2.head(2)
```

```python
data2.sex = encoded_sex
data2.smoker = encoded_smoker
data2.day = encoded_day
data2.time = encoded_time
```

```python
data2.head(2)
```

```python
X = data2[['total_bill', 'sex', 'smoker', 'day', 'time', 'size']].values
Y = data2['tip'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y , test_size=0.2, 
                                                random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
predicted_y = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, predicted_y)
r2 = r2_score(y_test,predicted_y)

print(mse)
print(r2)
```
