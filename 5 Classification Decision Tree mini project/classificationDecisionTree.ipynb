```python
import numpy as np
import matplotlib as plt 
import pandas as pd
```

```python
data = pd.read_csv('drug200.csv')
data.head(5)
```

```python
# checking for duplicateds and nulls 

data.isnull().sum()
data.duplicated().sum()
```

```python
data.columns
```

```python
# converting data frame to numpy array

input_feachers = data[['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']].values
input_labels = data['Drug']                     
```

```python
# converting non numeric values to numeric ones 

from sklearn.preprocessing import LabelEncoder

le_sex = LabelEncoder()
le_sex.fit(['F','M'])
input_feachers[:,1] = le_sex.transform(input_feachers[:,1]) 


le_BP = LabelEncoder()
le_BP.fit([ 'LOW', 'NORMAL', 'HIGH'])
input_feachers[:,2] = le_BP.transform(input_feachers[:,2])


le_Chol = LabelEncoder()
le_Chol.fit([ 'NORMAL', 'HIGH'])
input_feachers[:,3] = le_Chol.transform(input_feachers[:,3]) 

input_feachers[0:5]

```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input_feachers, input_labels, train_size=.8)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
y_test[0:3]
```

```python
y_predict[0:3]
```

```python
# evaluation

from sklearn.metrics import accuracy_score
accuracy_score(y_train, model.predict(x_train))
```

```python
accuracy_score(y_test, y_predict)
```
